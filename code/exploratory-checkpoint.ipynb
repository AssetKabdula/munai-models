```python
from openpyxl import load_workbook

def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

```python
file_sheet_dict = {
    
}
import re

p = re.compile(r'^.*(152|154).*$')
for file in os.listdir("../data/raw"):
    print(file)
    lst = get_sheetnames_xlsx(os.path.join(parent_dir, file))
    flst = [name for name in lst if p.match(name)]
    print("\t\t", flst) 
    if flst:
        file_sheet_dict[file] = flst[0]
file_sheet_dict
```
